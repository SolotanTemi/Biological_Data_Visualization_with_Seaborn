```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid')
print('Seaborn imported successfully')
```

```python
# Load the bacteria dataset
bacteria_path = 'https://raw.githubusercontent.com/HackBio-Internship/2025_project_collection/main/bacteria.csv'
bacteria_data = pd.read_csv(bacteria_path)

bacteria_data.head()
```

## PART 1: DISTRIBUTION PLOTS
These plots show the distribution and frequency of fitness scores in the bacteria dataset. Grouping by phenotype allows us to observe whether the phenotypic states differ in their antibiotic fitness profiles.

```python
# A histogram showing the distribution of carb_fit values
sns.histplot(data=bacteria_data, x='carb_fit', bins=20)  # use carb_fit as the x variable
plt.title('Distribution of Carbapenem fitness')              # add a descriptive title
plt.xlabel('carb_fit')                                    # label the x-axis
plt.ylabel('Count')                                       # label the y-axis
plt.show()                                                # display the plot
```

```python
# A smoothed density (KDE) plot of cipro_fit
sns.kdeplot(data=bacteria_data, x='cipro_fit', fill=True) 
plt.title('Density of Ciprofloxacin Fitness')                  
plt.xlabel('cipro_fit')                                   
plt.ylabel('Density')                                     
plt.show()                                               
```

```python
# Use displot to compare carb_fit distributions across Phenotype groups
sns.displot(
    data=bacteria_data,        # use the bacteria_data DataFrame
    x='carb_fit',              # plot carb_fit on the x-axis
    hue='Phenotype',           # colour the distributions by Phenotype
    kind='hist',               # use histogram as the underlying plot type
    bins=20,                   # choose the number of bins for the histogram
    multiple='stack'           # how overlapping bars are displayed ('stack', 'layer', or 'dodge')
)
plt.xlabel('carb_fit')        
plt.ylabel('Count')            
plt.title('Distribution of Carbapenem Fitness grouped by Phenotype')  
plt.show()                     
```

## PART 2: RELATIONSHIP PLOTS

These plots compare antibiotic fitness scores against each other, coloured by different categorical variables. This allows us to assess whether group membership influences the relationship between fitness traits.

```python
# Scatter plot of carb_fit vs pip_fit colored by labels
sns.scatterplot(data=bacteria_data, x='carb_fit', y='pip_fit', hue='labels')
plt.title('Carbapenem Fitness vs Piperacillin Fitness by labels')
plt.xlabel('carb_fit')
plt.ylabel('pip_fit')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()                                  

# Scatter plot of carb_fit vs pip_fit colored by Phenotype
sns.scatterplot(data=bacteria_data, x='carb_fit', y='pip_fit', hue='Phenotype')
plt.title('Carbapenem Fitness vs Piperacillin Fitness by Phenotype')
plt.xlabel('carb_fit')
plt.ylabel('pip_fit')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()    

# Scatter plot of carb_fit vs pip_fit colored by Isolation origin
sns.scatterplot(data=bacteria_data, x='carb_fit', y='pip_fit', hue='Isolation origin')
plt.title('Carbapenem Fitness vs Piperacillin Fitness by Isolation origin')
plt.xlabel('carb_fit')
plt.ylabel('pip_fit')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()                                  
```

```python
# A line plot of kan_fit vs genta_fit, colored by labels
sns.lineplot(
    data=bacteria_data,      # use the bacteria_data DataFrame
    x='kan_fit',             # plot kan_fit on the x-axis
    y='genta_fit',           # plot genta_fit on the y-axis
    hue='labels',            # colour the lines by labels
    marker='o'               # show markers at each data point
)
plt.title('Kanamycin Fitness vs Gentamicin Fitness by labels')
plt.xlabel('kan_fit')
plt.ylabel('genta_fit')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# A line plot of kan_fit vs genta_fit, colored by Phenotype
sns.lineplot(data=bacteria_data, x='kan_fit', y='genta_fit', hue='Phenotype', marker='o')
plt.title('Kanamycin Fitness vs Gentamicin Fitness by Phenotype')
plt.xlabel('kan_fit')
plt.ylabel('genta_fit')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# A line plot of kan_fit vs genta_fit, colored by Isolation origin
sns.lineplot(data=bacteria_data, x='kan_fit', y='genta_fit', hue='Isolation origin', marker='o')
plt.title('Kanamycin Fitness vs Gentamicin Fitness by Isolation origin')
plt.xlabel('kan_fit')
plt.ylabel('genta_fit')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')
bacteria_path = 'https://raw.githubusercontent.com/HackBio-Internship/2025_project_collection/main/bacteria.csv'
bacteria_data = pd.read_csv(bacteria_path)

# Pair plot of bacteria_data colored by labels

sns.pairplot(bacteria_data,   # use the full bacteria dataset
    hue='labels')         # color points by the labels column
plt.suptitle('Pairwise relationships between numeric features (colored by labels)')
plt.show()
```

## PART 4: HEATMAP OF FITNESS VALUES
The heatmap shows the degree of association between antibiotic fitness measurements across all bacterial isolates. A value approaching +1.0 reflects strong co-directional fitness responses, potentially indicative of shared or cross-resistance mechanisms, while values near 0 suggest independent resistance profiles.

```python
# Creating a heatmap using the numerical columns that represent fitness measurements.
# Isolating only the numerical fitness columns
fitness_cols = [col for col in bacteria_data.columns if col.endswith('_fit')]

# Building a matrix containing only those fitness measurements
fitness_matrix = bacteria_data[fitness_cols]

# Calculating the pairwise correlation between all fitness columns
# Values range from -1 (negative correlation) to 1 (perfect positive correlation)
correlation_matrix = fitness_matrix.corr()

# Creating a heatmap of the fitness values with annotations
sns.heatmap(
    correlation_matrix,  # rows = fitness variables, columns = fitness variables
    annot=True,          # show the numerical fitness values inside each cell
    fmt=".2f",           # format numbers to 2 decimal places for readability
    cmap='coolwarm'      # use the colourmap style
)
plt.title('Correlation Between Antibiotic Fitness Scores')
plt.xlabel('Fitness Measurements')      
plt.ylabel('Fitness Measurements')  
plt.tight_layout()
plt.show()
```

## PART 5: FACETED CATEGORICAL PLOT
Faceting slices the data dynamically. Instead of a crowded single plot, the relationship between fitness scores and other variables is displayed in separate panels. This makes it easier to compare fitness patterns across conditions without the visual clutter of a single combined plot.

```python
# Using catplot to make box plot
# Box plot of carb_fit by Phenotype, split by Isolation origin
sns.catplot(
    data=bacteria_data,               # bacteria dataset
    x='Phenotype',                    # main grouping on the x-axis
    y='carb_fit',                     # numeric fitness variable
    hue='Isolation origin',           # colour boxes by Isolation origin
    kind='box'                        # use box plots
)
plt.title('Carbapenem Fitness distribution by Phenotype and Isolation origin')
plt.xlabel('Phenotype')
plt.ylabel('carb_fit')
plt.show()
```

```python
# Using catplot to make box plot
# Box plots of genta_fit by labels, separated by Phenotype
sns.catplot(data=bacteria_data, x='labels', y='genta_fit', col='Phenotype', kind='box')
plt.suptitle('Gentamicin Fitness by Labels, Split by Phenotype', y=1.05)
plt.xlabel('labels')
plt.ylabel('genta_fit')
plt.show()
```
